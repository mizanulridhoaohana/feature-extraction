# Exploratory Data Analytics

### Import Libraries

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Show the image path
corrosion = 'data/CORROSION/'
non_corrosion = 'data/NOCORROSION/'
new_dir = 'split/'
```

```python
# Create objects that stores all the relevant images
sum_corrosion = [file for file in os.listdir(corrosion) if file.endswith('.jpg')]
sum_non_corrosion = [file for file in os.listdir(non_corrosion) if file.endswith('.jpg')]

print('There are', len(sum_corrosion), 'NORMAL images')
print('There are', len(sum_non_corrosion), 'CORROSION images')
```

```python
#Plot number of classes to identify imbalances
number_classes = {'Corrosion':len(sum_corrosion),
                  'Non-Corrosion':len(sum_non_corrosion)}

labels = number_classes.keys()
sizes = number_classes.values()

plt.bar(labels, sizes, width = 0.5)
plt.title("Number of images by Class")
plt.xlabel("Class Name")
plt.ylabel("Numer of Images")
```

```python
# Create a list of total images per label
total_images = [f'Total: {size}' for size in sizes]
explode = [0, 0.1]

# Create a pie chart with labels and percentages
plt.figure(figsize=(12, 12))
plt.pie(sizes, explode=explode, autopct='%3.1f%%', shadow=True, labels=labels, colors=('turquoise', 'orange'))
plt.title('Corrosion vs Non-Corrosion', fontsize=28)
plt.show()
```

### Chek Image

```python
corrotion_dir = 'data/CORROSION/'
non_corrosion_dir = 'data/NOCORROSION/'


plt.figure(figsize = (20,10))
num_imgs_1 = 4
for index in range(num_imgs_1):
    train_corrosion_pic_name = os.listdir(corrotion_dir)[index+10]
    train_corrosion_pic_address = corrotion_dir + train_corrosion_pic_name
#    fig, ax_list = plt.sunplots(nrows = 2, ncols =4)
    plt.subplot(2, num_imgs_1, index + 1)
    plt.imshow(Image.open(train_corrosion_pic_address))
    plt.tick_params(axis='both', which='both', bottom=False, top=False, labelbottom=False, labelleft=False)
    plt.title('Corrosion')

    train_nocorrosion_pic_name = os.listdir(non_corrosion_dir)[index+10]
    train_nocorrosion_pic_address = non_corrosion_dir + train_nocorrosion_pic_name

    plt.subplot(2, num_imgs_1, index + num_imgs_1 + 1)
    plt.imshow(Image.open(train_nocorrosion_pic_address))
    plt.tick_params(axis='both', which='both', bottom=False, top=False, labelbottom=False, labelleft=False)
    plt.title('No Corrosion')
```

### Check Image Size

```python
num_imgs_2 = 15
for index in range(num_imgs_2):
    train_corrosion_pic_name = os.listdir(corrotion_dir)[index]
    train_corrosion_pic_address = corrotion_dir + train_corrosion_pic_name
    train_corrosion_image = Image.open(train_corrosion_pic_address)
    train_corrosion_width, train_corrosion_height = train_corrosion_image.size
    print('train_CORROSION_image', index + 1, ":" , train_corrosion_width, 'x', train_corrosion_height)

    train_nocorrosion_pic_name = os.listdir(non_corrosion_dir)[index]
    train_nocorrosion_pic_address = non_corrosion_dir + train_nocorrosion_pic_name
    train_nocorrosion_image = Image.open(train_nocorrosion_pic_address)
    train_nocorrosion_width, train_nocorrosion_height = train_nocorrosion_image.size
    print('train_NOCORROSION_image', index + 1, ":" , train_nocorrosion_width, 'x', train_nocorrosion_height)
    print('\n')
```
